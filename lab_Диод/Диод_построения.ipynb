```python
from matplotlib.patches import Rectangle as rect

def plot_error(x_m, y_m, x_err, y_err, color="red", title="ВАХ",  x_label="В", y_label="mА", mod=1):
    #plot(x_m, y_m)
    fig=plt.figure()
    ax=fig.add_subplot()
    n=len(x_m)
    plot_and_approx(x_m, y_m, title=title, x_label=x_label, y_label=y_label, mod=mod)
    for i in range(n):
        x=x_m[i]
        y=y_m[i]
        x_e=x_err[i]
        y_e=y_err[i]
        x_n=x-x_e
        y_n=y-y_e
        ax.add_patch(rect((x_n, y_n), x_e*2., y_e*2., edgecolor=color,fc="none"))
    plt.show()
```

# Построение прямой Ричардсона

Прямая Ричардсона
ln Jₛ − 2 ln T = ln( Sₐ ⋅ A) − e Φ/kT
   или  
   ln Jₛ − 2 ln T = f(1/T)



 𝑇(𝑖нак ) = 970.52 + 996.208 · 𝑖нак

```python
def temp(i_n):
  return 970.52+996.208*i_n
```

```python
def plot_Richardson(ia, i_n, color="blue", title="ВАХ", x_label="A", y_label="B"):
  ia=np.array(ia)
  i_n=np.array(i_n)
  y_ax=[np.log(ia[i])-2*np.log(temp(i_n[i])) for i in range(len(ia))]
  x_ax=1/temp(i_n)
  plot(x_ax, y_ax, color=color, title="Прямая Ричардсона",  y_label="ln Jₛ − 2 ln T", x_label="1/T")
```

```python
plot_Richardson(ia1, in1)
plot_Richardson(ia2, in2, color="g")
plot_Richardson(ia3, in3, color="r")
```
